# Creating 3 different sheets to compute the Total Units and Total Volume Sold

```python
import pandas as pd
sheet1 = pd.read_csv('All_Channel_Sales_Units.csv')
sheet1
```

```python
sheet = sheet1.drop(sheet1[sheet1.Units <= 0.0].index)
```

```python
sheet
```

```python
sheet.to_excel('All_Channel_Sales_Units_modified.xlsx', sheet_name='sheet11', index=False)
```

```python
import pandas as pd
sheet2 = pd.read_csv('All_Channel_Sales_Volume.csv')
sheet2 = sheet2.drop(sheet2[sheet2.Volume <= 0.0].index)
sheet2
```

```python
sheet2.to_excel('All_Channel_Sales_Volume_modified.xlsx', sheet_name='sheet12', index=False)
```

```python
sheet3 = pd.merge(sheet, sheet2, on='Sku', how='outer')
```

```python
sheet3 = pd.DataFrame(sheet3.drop(['Year_x', 'Channel_x', 'Brand_Flag_x' , 'Year_y' , 'Channel_y' , 'Merged Sku_y','Month_y','Brand_Flag_y' ], axis = 1))
```

```python
sheet3.isnull().sum()
sheet3.Volume.fillna(value = 1,
          inplace = True)
sheet3.Units.fillna(value = 1,
          inplace = True)
# sheet3.drop
```

```python
sheet3.dropna(axis = 0,inplace = True)
```

```python
sheet3.isnull().sum()
```

```python
Total_Products_Sold = list(sheet3.Units)
units = list(sheet3.Units)
Volume = list(sheet3.Volume)
for i in range(len(Total_Products_Sold)):
    Total_Products_Sold[i] = round((units[i]*Volume[i]),3)
#     Total_Products_Sold.append(round(Tot,3))
    
```

```python
Total_Products_Sold
```

```python
sheet3['Total Products Sold'] = Total_Products_Sold
```

```python
sheet3
```

```python
sheet3.isnull().sum()
```

```python
sheet3.to_excel('All_Channel_Tot_Sales_made_modified.xlsx', sheet_name='sheet13', index=False)
```
